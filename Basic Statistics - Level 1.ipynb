# Q9) :
- ***Calculate Skewness, Kurtosis & draw inferences on the following data Cars.***
- **speed and distance** 
    - Use Q9_a.csv
- **SP and Weight(WT)**
    - Use Q9_b.csv


```python
# Loading "Q9_a.csv" dataset
df_a = pd.read_csv('Q9_a.csv',index_col=0)

# Showing dataset
df_a
```

```python
# Calculate the skewness and kurtosis of the 'speed' and 'distance' columns
speed_skewness = df_a['speed'].skew()
speed_kurtosis = df_a['speed'].kurtosis()
dist_skewness = df_a['dist'].skew()
dist_kurtosis = df_a['dist'].kurtosis()

# Plot the distribution of the 'speed' and 'distance' columns with kdeplot
sns.kdeplot(df_a['speed'], label=f"speed (Skewness = {speed_skewness:.2f}, Kurtosis = {speed_kurtosis:.2f})")
sns.kdeplot(df_a['dist'], label=f"dist (Skewness = {dist_skewness:.2f}, Kurtosis = {dist_kurtosis:.2f})")

# Add a title and label to the plot
plt.title("Kernel Density Estimation of speed and dist Columns")
plt.xlabel("Value")
plt.legend(loc='center right')

# Show the plot
plt.show()
```

```python
# Loading "Q9_b.csv" dataset
df_b = pd.read_csv('Q9_b.csv',index_col=0)

# Showing dataset
df_b
```

```python
# Calculate the skewness and kurtosis of the 'speed' and 'distance' columns
SP_skewness = df_b['SP'].skew()
SP_kurtosis = df_b['SP'].kurtosis()
WT_skewness = df_b['WT'].skew()
WT_kurtosis = df_b['WT'].kurtosis()

# Plot the distribution of the 'speed' and 'distance' columns with kdeplot
sns.kdeplot(df_b['SP'], label=f"SP (Skewness = {SP_skewness:.2f}, Kurtosis = {SP_kurtosis:.2f})")
sns.kdeplot(df_b['WT'], label=f"WT (Skewness = {WT_skewness:.2f}, Kurtosis = {WT_kurtosis:.2f})")

# Add a title and label to the plot
plt.title("Kernel Density Estimation of SP and WT Columns")
plt.xlabel("Value")
plt.legend()

# Show the plot
plt.show()
```

# Q. 20 :
***Calculate probability from the given dataset for the below cases.***
- ***Data _set: Cars.csv***

- **Calculate the probability of MPG  of Cars for the below cases.**
    - a) P(MPG>38)
    - b) P(MPG<40)
    - c) P(20<MPG<50)

```python
# Loading dataset
df_car = pd.read_csv("Cars.csv")

# Showing dataset
df_car
```

```python
# Extracting the MPG column from dataset
mpg = df_car["MPG"]
```

```python
# a. P(MPG > 38)
prob_a = 1 - stats.norm.cdf(38, mpg.mean(), mpg.std())
print("P(MPG > 38) =", prob_a)

# b. P(MPG < 40)
prob_b = stats.norm.cdf(40, mpg.mean(), mpg.std())
print("P(MPG < 40) =", prob_b)

# c. P(20 < MPG < 50)
prob_c = stats.norm.cdf(50, mpg.mean(), mpg.std()) - stats.norm.cdf(20, mpg.mean(), mpg.std())
print("P(20 < MPG < 50) =", prob_c)
```

# Q. 22 :
***Calculate the Z scores of 90% confidence interval, 94% confidence interval, 60% confidence interval.***

```python
confidence_interval_90 = 0.90
z_score_90 = stats.norm.ppf(confidence_interval_90)
print("Z-score for 90% confidence interval:", z_score_90)

confidence_interval_94 = 0.94
z_score_94 = stats.norm.ppf(confidence_interval_94)
print("Z-score for 94% confidence interval:", z_score_94)

confidence_interval_60 = 0.60
z_score_60 = stats.norm.ppf(confidence_interval_60)
print("Z-score for 60% confidence interval:", z_score_60)
```

# Q. 23 :
***Calculate the t scores of 95% confidence interval, 96% confidence interval, 99% confidence interval for sample size of 25***

```python
sample_size = 25
degrees_of_freedom = sample_size - 1

confidence_interval_95 = 0.95
t_score_95 = stats.t.ppf(confidence_interval_95, degrees_of_freedom)
print("t-score for 95% confidence interval:", t_score_95)

confidence_interval_96 = 0.96
t_score_96 = stats.t.ppf(confidence_interval_96, degrees_of_freedom)
print("t-score for 96% confidence interval:", t_score_96)

confidence_interval_99 = 0.99
t_score_99 = stats.t.ppf(confidence_interval_99, degrees_of_freedom)
print("t-score for 99% confidence interval:", t_score_99)
```

# Q. 24 :
***A Government  company claims that an average light bulb lasts 270 days. A researcher randomly selects 18 bulbs for testing. The sampled bulbs last an average of 260 days, with a standard deviation of 90 days. If the CEO's claim were true, what is the probability that 18 randomly selected bulbs would have an average life of no more than 260 days.***

```python
mean = 270
n = 18
x_bar = 260
s = 90
df = n - 1

standard_error = s / (n ** 0.5)
t_score = (x_bar - mean) / standard_error
p = stats.t.cdf(t_score, df)

print("The probability that the average life of 18 randomly selected bulbs would be no more than 260 days is:", p)
```
